```python
# Leer un csvw
df = pd.read_csv('../data/CompleteDataset.csv')
print('Dataaset COMPLETO',df.columns)
print('Longitud de la consulta', len(df))


```

```python
# Eliminar espacios en blanco al principio y al final en todo el DataFrame
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
print(df.info())
```

```python
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtrar las columnas problemáticas
columns_to_modify = [
    'Acceleration', 
    'Aggression', 
    'Agility', 
    'Balance', 
    'Ball control', 
    'Composure', 
    'Crossing', 
    'Curve', 
    'Dribbling',
    'Free kick accuracy',
    'GK diving',
    'GK handling',
    'GK kicking',
    'GK positioning',
    'GK reflexes',
    'Heading accuracy',
    'Interceptions',
    'Jumping',
    'Long passing',
    'Long shots',
    'Marking',
    'Penalties',
    'Positioning',
    'Reactions',
    'Short passing',
    'Shot power',
    'Sliding tackle',
    'Sprint speed',
    'Stamina',
    'Standing tackle',
    'Strength',
    'Vision',
    'Volleys'
]

# Concatenar todos los valores únicos en un solo dataframe
unique_values = pd.concat([df[col].astype(str) for col in columns_to_modify], axis=0).value_counts()

# Filtrar los valores que tienen un formato extraño
atypical_values = unique_values[unique_values.index.str.contains(r'[+\-]', na=False)]

# Seleccionar un subconjunto de valores atípicos para la gráfica (por ejemplo, los 20 más comunes)
subset_atypical_values = atypical_values.head(20)

# Mostrar los valores atípicos seleccionados y su frecuencia
print(subset_atypical_values)
```

```python
# Crear el gráfico de barras
plt.figure(figsize=(14, 7))
subset_atypical_values.plot(kind='bar')
plt.title('Frecuencia de algunos valores atípicos en columnas seleccionadas')
plt.xlabel('Valores únicos')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90)
plt.show()
```

```python
subset_atypical_values
```

```python

sample_size = 20  

def parteEntera(value):
    if isinstance(value, str):
        if '+' in value or '-' in value:
            return value.split('+')[0].split('-')[0]
    return value

df_cleaned = df.copy()
for column in columns_to_modify:
    df_cleaned[column] = df_cleaned[column].apply(parteEntera)

df_cleaned[columns_to_modify] = df_cleaned[columns_to_modify].apply(pd.to_numeric, errors='coerce')

def top_values_before_cleaning(column_name, df_original):
    plt.figure(figsize=(10, 6))
    top_values_original = df_original[column_name].value_counts().head(sample_size)
    top_values_original.sort_index().plot(kind='bar')
    plt.title(f'{column_name} - Valores más frecuentes antes de la limpieza')
    plt.xlabel('Valores')
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

columns_to_show = columns_to_modify[:2] 
for column in columns_to_show:
    top_values_before_cleaning(column, df)
```

```python
for column in columns_to_modify:
    df[column] = df[column].apply(parteEntera)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calcular el número de valores nulos por cada columna
null_counts = df_cleaned.isnull().sum()

# Filtrar las columnas que tienen valores nulos
columns_with_nulls = null_counts[null_counts > 0]

# Mostrar los números de valores nulos en una tabla
null_counts_table = pd.DataFrame(columns_with_nulls)
null_counts_table.columns = ['Cantidad de nulos']
null_counts_table.index.name = 'Columnas con nulos'

# Mostrar la tabla
print(null_counts_table)

# Crear un gráfico de pie con los valores absolutos de nulos
plt.figure(figsize=(8, 8))
plt.pie(columns_with_nulls, labels=columns_with_nulls.index, autopct=lambda pct: f'{int(pct * sum(columns_with_nulls) / 100)}')
plt.title('Cantidad de valores nulos por columna')
plt.axis('equal')  # Asegura que el gráfico sea un círculo
plt.show()
```

## al imputar los valores faltantes con la media en estas columnas específicas, estamos asegurando que nuestro análisis y cualquier visualización o dashboard que construyamos reflejen con precisión las capacidades promedio de los jugadores en esos atributos clave. Esto contribuye a una interpretación más informada y confiable de los datos para cualquier usuario interesado en el análisis de futbolistas.

```python
from sklearn.impute import SimpleImputer

# Seleccionar las columnas que tienen valores nulos
columns_to_impute = [
    'CAM', 'CB', 'CDM', 'CF', 'CM', 'LAM', 'LB', 'LCB', 'LCM', 
    'LDM', 'LF', 'LM', 'LS', 'LW', 'LWB', 'RAM', 'RB', 'RCB', 
    'RCM', 'RDM', 'RF', 'RM', 'RS', 'RW', 'RWB', 'ST', 'Free kick accuracy', 'Penalties'
]

# Crear un imputador para llenar los valores nulos con la media
imputer = SimpleImputer(strategy='mean')

# Aplicar el imputador a las columnas seleccionadas
df_cleaned[columns_to_impute] = imputer.fit_transform(df_cleaned[columns_to_impute])
```
