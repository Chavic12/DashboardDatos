```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtrar las columnas problemáticas
columns_to_modify = [
    'Acceleration', 
    'Aggression', 
    'Agility', 
    'Balance', 
    'Ball control', 
    'Composure', 
    'Crossing', 
    'Curve', 
    'Dribbling',
    'Free kick accuracy',
    'GK diving',
    'GK handling',
    'GK kicking',
    'GK positioning',
    'GK reflexes',
    'Heading accuracy',
    'Interceptions',
    'Jumping',
    'Long passing',
    'Long shots',
    'Marking',
    'Penalties',
    'Positioning',
    'Reactions',
    'Short passing',
    'Shot power',
    'Sliding tackle',
    'Sprint speed',
    'Stamina',
    'Standing tackle',
    'Strength',
    'Vision',
    'Volleys'
]

# Concatenar todos los valores únicos en un solo dataframe
unique_values = pd.concat([df[col].astype(str) for col in columns_to_modify], axis=0).value_counts()

# Filtrar los valores que tienen un formato extraño
atypical_values = unique_values[unique_values.index.str.contains(r'[+\-]', na=False)]

# Seleccionar un subconjunto de valores atípicos para la gráfica (por ejemplo, los 20 más comunes)
subset_atypical_values = atypical_values.head(20)

# Mostrar los valores atípicos seleccionados y su frecuencia
print(subset_atypical_values)
```

```python
# Crear el gráfico de barras
plt.figure(figsize=(14, 7))
subset_atypical_values.plot(kind='bar')
plt.title('Frecuencia de algunos valores atípicos en columnas seleccionadas')
plt.xlabel('Valores únicos')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90)
plt.show()
```

```python

sample_size = 20  

def parteEntera(value):
    if isinstance(value, str):
        if '+' in value or '-' in value:
            return value.split('+')[0].split('-')[0]
    return value

df_cleaned = df.copy()
for column in columns_to_modify:
    df_cleaned[column] = df_cleaned[column].apply(parteEntera)

df_cleaned[columns_to_modify] = df_cleaned[columns_to_modify].apply(pd.to_numeric, errors='coerce')

def top_values_before_cleaning(column_name, df_original):
    plt.figure(figsize=(10, 6))
    top_values_original = df_original[column_name].value_counts().head(sample_size)
    top_values_original.sort_index().plot(kind='bar')
    plt.title(f'{column_name} - Valores más frecuentes antes de la limpieza')
    plt.xlabel('Valores')
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

columns_to_show = columns_to_modify[:2] 
for column in columns_to_show:
    top_values_before_cleaning(column, df)
```
